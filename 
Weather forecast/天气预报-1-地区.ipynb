```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.110 Safari/537.36',
}

url = "https://m.tianqi.com/lishi/shanghai"
html = requests.get(url).text
```

### 解析网页

```python
soup1 = BeautifulSoup(html, 'html.parser')
```

### 获取网页指定内容


### 三种方法

```python
## 方法一：find 找到class为newSearchList的div，再find_all 下面的所有a标签
# res = soup1.find(class_="newSearchList").find_all('a')

## 方法二：find 找到class为newSearchList的div，再select 所有 ul li 下的 a 标签
res = soup1.find(class_="newSearchList").select('ul li a')

## 获取 a 标签内的 href title 属性，标签内的文本

for url in res:
    print(url.get('href'))
    print(url.get('title'))
    print(url.get_text())
```
